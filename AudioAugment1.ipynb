```python
import os
import numpy as np
import librosa
import librosa.display
import soundfile as sf
import random
from audiomentations import Compose, AddGaussianNoise, PitchShift, TimeStretch, Shift
```

```python
#Apply augmentation

class_dirs={
    "belly_pain": "/content/drive/MyDrive/Audio Files infant cry zip/donateacry_corpus/belly_pain",
    "burping": "/content/drive/MyDrive/Audio Files infant cry zip/donateacry_corpus/burping",
    "discomfort": "/content/drive/MyDrive/Audio Files infant cry zip/donateacry_corpus/discomfort",
    "hungry": "/content/drive/MyDrive/Audio Files infant cry zip/donateacry_corpus/hungry",
    "tired": "/content/drive/MyDrive/Audio Files infant cry zip/donateacry_corpus/tired"
}
```

```python
augment=Compose([
    AddGaussianNoise(min_amplitude=0.001, max_amplitude=0.015, p=0.8),
    PitchShift(min_semitones=-2, max_semitones=2, p=0.8),
    TimeStretch(min_rate=0.8,max_rate=1.25,p=0.8),
    Shift(min_shift=-0.2, max_shift=0.2,p=0.8)
])
```

```python
def augment_audio(file_path, output_dir, class_name, aug_index):
    """Applies augmentation to an audio file and saves it."""
    y, sr = librosa.load(file_path, sr=None)

    augmented_audio = augment(samples=y, sample_rate=sr)

    # Define output file path in Google Drive
    output_path = os.path.join(output_dir, class_name, f"{os.path.basename(file_path).split('.')[0]}_aug_{aug_index}.wav")
    os.makedirs(os.path.dirname(output_path), exist_ok=True)

    # Save the augmented audio file
    sf.write(output_path, augmented_audio, sr)

# Step 6: Function to generate augmented samples for each class
def process_class(input_dir, output_dir, class_name, target_count):
    """Generates augmented audio files to ensure each class has exactly 'target_count' samples."""
    os.makedirs(os.path.join(output_dir, class_name), exist_ok=True)

    # Get all existing audio files in the class folder
    files = [os.path.join(input_dir, f) for f in os.listdir(input_dir) if f.endswith('.wav')]

    # If the number of samples is already 200, skip augmentation
    if len(files) >= target_count:
        print(f" {class_name} already has {len(files)} files. No augmentation needed.")
        return

    # Compute how many more samples are needed
    missing_samples = target_count - len(files)

    aug_index = 1  # Track augmented file numbering

    while missing_samples > 0:
        file_to_augment = random.choice(files)
        augment_audio(file_to_augment, output_dir, class_name, aug_index)
        aug_index += 1
        missing_samples -= 1  # Reduce the remaining count

    print(f" {class_name} augmented to {target_count} samples!")

# Step 7: Process all classes and apply augmentation
for class_name, path in class_dirs.items():
    class_output = os.path.join(drive_output_dir)

    # Set target samples (200 for all classes, including undersampling "hungry")
    target_samples = 200 if class_name != "hungry" else 200

    process_class(path, class_output, class_name, target_samples)

print(" Augmentation completed! Files are saved in Google Drive.")
```
